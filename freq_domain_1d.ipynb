Convert processed freq domain file to np array

```python
import numpy as np
import matplotlib.pyplot as plt    
```

Bruker uses little endian for raw files

```python
folder = 'nmr_data/2'
```

```python
dtype = '<i4'
def data_gen(in_r, in_i):
    with open(in_r, 'rb') as input_serial_file: 
        raw_data_r = np.frombuffer(input_serial_file.read(), dtype=dtype)
    with open(in_i, 'rb') as input_serial_file:
        raw_data_i = np.frombuffer(input_serial_file.read(), dtype=dtype)

    raw_data = raw_data_r + 1j * raw_data_i
    return raw_data

freq_data = folder + '/pdata/1/'
r1 = freq_data + "1r"
i1 = freq_data + "1i"
signals = data_gen(r1, i1)
print(signals.shape)
```

```python
plt.figure(figsize=(8, 5), dpi=300)                   
plt.subplot(2, 1, 1)   
plt.plot(np.real(signals), c='r', label='real')  
plt.legend()                                 

plt.subplot(2, 1, 2)                  
plt.plot(np.imag(signals),c='k', label='imaginary')
plt.legend()                               

plt.show()   
```
